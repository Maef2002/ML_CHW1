## Array creation
>Create the followings to numpy array:
* A: a 3\*3 array like this:
$$
A = \left(\begin{array}{cc} 
10 & 20 & 30\\
5 & 10 & 15\\
1 & 5 & 8
\end{array}\right)
$$ 
* B: a 3\*3\*4 that filled with the numbers start from 15 and have steps of 5 to the end. The array should be filled first along the first dimension then the second dimension and at last the third dimension. like this:
$$
B[:, :, 0] = \left(\begin{array}{cc} 
15 & 30 & 45\\
20 & 35 & 50\\
25 & 40 & 55
\end{array}\right)
$$ 

```python
### START CODE HERE ###
A = np.array([[10, 20, 30], [5, 10, 15], [1, 5, 8]])
B = np.zeros((3,3,4,3,3))
zero = 15
for i in range(3):
    for j in range(3):
        for k in range(4):
            B[i][j][k] = np.array([[zero, zero+5, zero+10],[zero+15, zero+20, zero+25],[zero+30, zero+35, zero+40]])
            zero += 45
            
print(A)            
print(B)
### END CODE HERE ###
```

## Plot
>You must plot the cos(x) curve. plot 20 cycle of the curve. the frequency of first half of curve is 50 Hz and frequency of second lalf is 200 Hz. 
> set the x_label, y_label and title for your plot.

```python
### START CODE HERE ###
from matplotlib.pyplot import figure

t = np.linspace(0,0.25,1000)
y = 0
for i in range(10):
    y += np.heaviside(t - i * 0.0125, 0.5) * np.heaviside(i * 0.010 - t, 0.5) * np.cos(2 * math.pi * 50)

y = np.heaviside(t, 1) * np.heaviside(0.2 - t, 1)  * np.cos(2 * math.pi * 50 * t) + np.heaviside(t-0.2, 1) * np.heaviside(0.25 - t, 1)  * np.cos(2 * math.pi * 200 * t)
figure(figsize=(10, 10))
plot.plot(t,y)
### END CODE HERE ###
```
